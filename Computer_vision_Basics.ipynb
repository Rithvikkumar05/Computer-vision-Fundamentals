```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

img = cv2.imread('q.jpg',0)
equalized_image = cv2.equalizeHist(img)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')


plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')
```

```python
import cv2
cap = cv2.VideoCapture('v1.mp4');
while (True):
  ret, frame=cap.read()
  gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
  cv2_imshow(gray)
  if cv2.waitKey(50000):
    break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
image = cv2.imread('app.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
resized_image = cv2.resize(image, (224, 224))  # Resize to 224x224 pixels
cv2_imshow(resized_image)
```

```python
nor = image / 255.0  # Normalize pixel values to [0, 1]
cv2_imshow(nor)
```

```python
blurred_image = cv2.GaussianBlur(image, (5, 5), 2.0)  # Apply Gaussian blur
cv2_imshow(blurred_image)
```

```python
_, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
cv2_imshow(binary_image)
print(_)
```

```python
edges = cv2.Canny(image, 100, 200)  # Canny edge detection
cv2_imshow(edges)
```

```python
line =cv2.line(image, (0, 00), (300, 300), (0, 0, 0), 2)
cv2_imshow(line)
```

```python
import datetime
font = cv2.FONT_HERSHEY_SIMPLEX
datet= str(datetime.datetime.now())
frame = cv2.putText(image, datet, (10, 50), font, 1, (0, 0, 255), 2, cv2.LINE_AA)
cv2_imshow( frame)
```
